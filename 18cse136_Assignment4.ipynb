## Data Visualization

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("toyota.csv")
data.head()
```

### Scatter Plot between Age and Price

```python
plt.scatter(data['Age'],data['Price'])
plt.xlabel('Age')
plt.ylabel('Price')
plt.show()
```

### Histogram of Data

```python
data.hist(figsize = (20,30))
```

### Bar Plots of Fuel Type

```python
counts = [979, 120, 12]
fuelType = ('Petrol', 'Diesel', 'CNG')
index = np.arange(len(fuelType))
plt.bar(index, counts, color=['red', 'blue', 'cyan'])
plt.title('Bar plot of fuel types')
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')
plt.xticks(index, fuelType, rotation = 90)
plt.show()
```

### Regression Plot

```python
sns.set(style="darkgrid")
sns.regplot(x=data['Age'],y=data['Price'])
```

### Distplot

```python
sns.distplot(data['Age'])
```

### Countplot

```python
sns.countplot(x="FuelType", data=data)
```

```python
sns.countplot(x="FuelType", data=data, hue = "Automatic")
```

### Boxplot

```python
sns.boxplot(y=data["Price"])
```

```python
sns.boxplot(x=data['FuelType'], y=data["Price"])
```

### Pairplot

```python
sns.pairplot(data, kind="countplot")
plt.show()
```
